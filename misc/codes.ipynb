```python
import pandas as pd 
import sqlite3 

def connect():
    return sqlite3.connect('py835/codes/codes.db')

def update_claim_adjustment_group_codes():
    conn = connect()
    claim_adjustment_group_codes = pd.read_csv('py835/codes/claim_adjustment_group_codes.csv').rename({'Code':'value', 'Description':'description'}, axis=1)
    fields = pd.DataFrame([{'field':'CAS01'} for _,row in claim_adjustment_group_codes.iterrows()])
    claim_adjustment_group_codes = pd.concat([fields, claim_adjustment_group_codes], axis=1)
    claim_adjustment_group_codes.to_sql('claim_adjustment_group_codes', conn, index=False, if_exists='replace')

def get_claim_adjustment_group_codes():
    conn = connect()
    return pd.read_sql_query("SELECT * FROM claim_adjustment_group_codes", conn)
def update_claim_adjustment_reason_codes():
    conn = connect()
    claim_adjustment_reason_codes = pd.read_csv('py835/codes/claim_adjustment_reason_codes.csv').rename({'Code':'value', 'Description':'description'}, axis=1)
    fields = pd.DataFrame([{'field':'CAS02'} for _,row in claim_adjustment_reason_codes.iterrows()])
    claim_adjustment_reason_codes = pd.concat([fields, claim_adjustment_reason_codes], axis=1)
    claim_adjustment_reason_codes.to_sql('claim_adjustment_reason_codes', conn, index=False, if_exists='replace')

def update_claim_adjustment_reason_codes():
    conn = connect()
    claim_adjustment_reason_codes = pd.read_csv('py835/codes/claim_adjustment_reason_codes.csv').rename({'Code':'value', 'Description':'description'}, axis=1)
    fields = ['CAS02','CAS05','CAS08','CAS11','CAS14','CAS17']
    dfs = []
    for field in fields:
        df = pd.DataFrame([{'field':field} for _,row in claim_adjustment_reason_codes.iterrows()])
        df = pd.concat([df, claim_adjustment_reason_codes], axis=1)
        dfs.append(df)
    claim_adjustment_reason_codes = pd.concat(dfs, axis=0)
    claim_adjustment_reason_codes.to_sql('claim_adjustment_reason_codes', conn, index=False, if_exists='replace')
def get_claim_adjustment_reason_codes():
    conn = connect()
    return pd.read_sql_query("SELECT * FROM claim_adjustment_reason_codes", conn)

def update_claim_status_category_codes():
    conn = connect()
    claim_status_category_codes = pd.read_csv('py835/codes/claim_status_category_codes.csv').rename({'Code':'value', 'Description':'description'}, axis=1)
    fields = []
    dfs = []
    for field in fields:
        df = pd.DataFrame([{'field':field} for _,row in claim_status_category_codes.iterrows()])
        df = pd.concat([df, claim_status_category_codes], axis=1)
        dfs.append(df)
    if dfs:
        claim_status_category_codes = pd.concat(dfs, axis=0)
        claim_status_category_codes.to_sql('claim_status_category_codes', conn, index=False, if_exists='replace')
def get_claim_status_category_codes():
    conn = connect()
    return pd.read_sql_query("SELECT * FROM claim_status_category_codes", conn)

def update_claim_status_codes():
    conn = connect()
    claim_status_codes = pd.read_csv('py835/codes/claim_status_codes.csv').rename({'Code':'value', 'Description':'description'}, axis=1)
    fields = ['CLP02']
    dfs = []
    for field in fields:
        df = pd.DataFrame([{'field':field} for _,row in claim_status_codes.iterrows()])
        df = pd.concat([df, claim_status_codes], axis=1)
        dfs.append(df)
    claim_status_codes = pd.concat(dfs, axis=0)
    claim_status_codes.to_sql('claim_status_codes', conn, index=False, if_exists='replace')
def get_claim_status_codes():
    conn = connect()
    return pd.read_sql_query("SELECT * FROM claim_status_codes", conn)

def update_error_reason_codes():
    conn = connect()
    error_reason_codes = pd.read_csv('py835/codes/error_reason_codes.csv').rename({'Code':'value', 'Description':'description'}, axis=1)
    fields = ['REF02']
    dfs = []
    for field in fields:
        df = pd.DataFrame([{'field':field} for _,row in error_reason_codes.iterrows()])
        df = pd.concat([df, error_reason_codes], axis=1)
        dfs.append(df)
    error_reason_codes = pd.concat(dfs, axis=0)
    error_reason_codes.to_sql('error_reason_codes', conn, index=False, if_exists='replace')
```
